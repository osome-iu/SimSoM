Test the implementation of Huberman decay of noveltty/recency r: $r \sim~ e^{-0.4t^{0.4}}$ where t is the time step

```python
import numpy as np
import matplotlib.pyplot as plt

ages = np.array([t for t in range(10)])
recency = np.exp(-0.4 * (ages**0.4))

aging_lambda = 0.9
lambdas = np.empty(len(ages), dtype=float)
lambdas.fill(aging_lambda)
old_recency = lambdas**ages

plt.plot(ages, recency, label="huberman")
plt.plot(ages, old_recency, label="exp 0.9")
plt.legend()
plt.xlabel("age")
plt.ylabel("recency")
```
